```python
def t_mean(t_max, t_min):
    result = np.where(t_max > 20 or t_min < -100, None, (t_max + t_min) / 2)
    
    return result
    
```

```python
array_t_min = t_min.air_temperature.to_numpy()
array_t_max = t_max.air_temperature.to_numpy()
```

```python
result = np.vectorize(t_mean)(array_t_max, array_t_min)
result
```
