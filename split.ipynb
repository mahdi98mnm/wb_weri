```python
import pandas as pd
import numpy as np
import xarray as xr
import pickle
from datetime import datetime


data_path = "D:\\mehdi\\DailyData\\Data\\"
```

```python
with open('P.pkl', 'rb') as f:
    p = pickle.load(f)
```

```python
def start_year(x):
    
    df = pd.DataFrame(
        {"Year" : np.datetime_as_string(x.time.values, unit='Y').tolist()}
    )
    
    df = df.drop_duplicates(keep='first')
    
    df.reset_index(drop=False, inplace=True)
    
    return df

def save_to_netcdf(data, data_path, para, year):
    
    data.to_netcdf(f"{data_path}{para}_{year}.nc")
```

```python
df_year = start_year(p)

for y, _ in enumerate(df_year["index"].to_list()):
        
    if y == len(df_year["index"].to_list()) - 1:
        break
    
    array_y = p.sel(time=slice(p.time.values[df_year["index"][y]], p.time.values[df_year["index"][y + 1] - 1]))
    
    save_to_netcdf(
        data=array_y,
        data_path=data_path,
        para="P",
        year=df_year["Year"][y]
    )
```
