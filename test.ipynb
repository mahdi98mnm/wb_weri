```python
import pickle
import glob
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib
```

```python
data_list_P = []
data_list_Temp = []
data_list_Tmax = []
data_list_Tmin = []
```

```python
import calendar

def cumulative_days(start_year, end_year):
    cumulative_days_list = [0]

    for year in range(start_year, end_year + 1):
        if calendar.isleap(year):
            days_in_year = 366
        else:
            days_in_year = 365

        cumulative_days_list.append(days_in_year)



    return cumulative_days_list

start_year = 1979
end_year = 2023
result = cumulative_days(start_year, end_year)


def calculate_cumulative_list(input_list):
    return [sum(input_list[:i+1]) for i in range(len(input_list))]

number_of_days = calculate_cumulative_list(result)
```

```python
pickle_files_P = glob.glob("P/*.pkl")

for file in pickle_files_P:
    with open(file, 'rb') as f:
        df = pd.read_pickle(f)
        data_list_P.append(df)

concatenated_dataset_P = xr.concat(data_list_P, dim='time')

with open(f'P.pkl', 'wb') as f:
    pickle.dump(concatenated_dataset_P, f)
```
